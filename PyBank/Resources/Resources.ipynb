```python
sum = df["Profit/Losses"].sum()
sum
print("Total: $",sum)
```

```python
avg_value=df["Profit/Losses"].mean()
avg_value
```

```python
print("Average Change :",avg_value)
```

```python
max_value_row=df.loc[df["Profit/Losses"].idxmax()]
max_value_row
```

```python
print("Greatest Increase in Profits:",max_value_row['Date'],max_value_row['Profit/Losses'])
```

```python
min_value_row=df.loc[df["Profit/Losses"].idxmin()]
min_value_row
```

```python
print("Greatest Decrease in Profits:",min_value_row['Date'],min_value_row['Profit/Losses'])
```
